```python
import os
import re
import glob
import torch
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageFile
from tqdm import tqdm
import torch.nn as nn
import torch.optim as optim
from torch.optim.lr_scheduler import ReduceLROnPlateau
import torchvision.models as models
import torchvision.transforms as transforms
from torch.utils.data import Dataset, DataLoader
import warnings

# Allow loading of truncated images
ImageFile.LOAD_TRUNCATED_IMAGES = True
Image.MAX_IMAGE_PIXELS = None

# ============================
# STEP 1: CREATE MATCHED DATASET
# ============================

def create_matched_dataset(csv_path, image_root):
    """Create dataset by matching CSV data with available image folders"""
    print("=== CREATING MATCHED DATASET ===")
   
    # Load CSV data
    df = pd.read_csv(csv_path)
    print(f"Original CSV: {len(df)} listings")
    print(f"Available columns: {list(df.columns)}")
   
    # Use existing ID column directly
    if 'id' not in df.columns:
        print("Error: 'id' column not found in CSV!")
        return None
   
    # Remove rows with missing IDs
    df_valid = df.dropna(subset=["id"]).copy()
    df_valid["listing_id"] = df_valid["id"].astype(str)
    print(f"Valid entries with IDs: {len(df_valid)}")
   
    # Get available image folders
    if not os.path.exists(image_root):
        print(f"Error: Image root directory {image_root} does not exist")
        return None
       
    available_folders = [d for d in os.listdir(image_root)
                        if os.path.isdir(os.path.join(image_root, d))]
    print(f"Available image folders: {len(available_folders)}")
   
    # Match listings with available folders
    df_valid["listing_id"] = df_valid["listing_id"].astype(str)
    available_folders_set = set(available_folders)
   
    matched_df = df_valid[df_valid["listing_id"].isin(available_folders_set)].copy()
    print(f"Matched listings: {len(matched_df)}")
   
    if len(matched_df) == 0:
        print("No matches found! Check if listing IDs match folder names.")
        return None
   
    # Create final dataset with listing ID and price
    final_dataset = pd.DataFrame({
        "listing_id": matched_df["listing_id"],
        "price": matched_df["price"]
    })
   
    # Remove any invalid prices
    final_dataset = final_dataset.dropna(subset=["price"])
    final_dataset = final_dataset[final_dataset["price"] > 0]

    # Clip prices to reasonable Airbnb range ($20-$300)
    print(f"Before price clipping: {len(final_dataset)} listings")
    final_dataset = final_dataset[
        (final_dataset["price"] >= 20) & 
        (final_dataset["price"] <= 300)
    ]
    print(f"After price clipping ($20-$300): {len(final_dataset)} listings")
   
    print(f"Final dataset: {len(final_dataset)} listings")
    print(f"Price range: ${final_dataset['price'].min():.2f} - ${final_dataset['price'].max():.2f}")
    print(f"Median price: ${final_dataset['price'].median():.2f}")
   
    return final_dataset

# ============================
# STEP 2: DATASET CLASS
# ============================

class AirbnbDataset(Dataset):
    def __init__(self, df, image_root, transform=None, max_images_per_listing=5):
        self.image_root = image_root
        self.transform = transform
        self.pairs = []
       
        print(f"Building dataset from {len(df)} listings...")
       
        # Create image-price pairs
        for _, row in tqdm(df.iterrows(), total=len(df), desc="Processing listings"):
            listing_id = str(row["listing_id"])
            price = float(row["price"])
           
            # Find images for this listing
            listing_dir = os.path.join(image_root, listing_id)
            if not os.path.exists(listing_dir):
                continue
               
            # Get all image files (prioritize jpg/jpeg for speed)
            images = glob.glob(os.path.join(listing_dir, "*.jpg"))
            if len(images) == 0:
                images = glob.glob(os.path.join(listing_dir, "*.jpeg"))
            if len(images) == 0:
                images.extend(glob.glob(os.path.join(listing_dir, "*.png")))
                images.extend(glob.glob(os.path.join(listing_dir, "*.webp")))
           
            if len(images) == 0:
                continue
               
            # Limit images per listing
            selected_images = images[:max_images_per_listing]
           
            # Add all selected images with their price
            for img_path in selected_images:
                try:
                    # Skip very large files (more aggressive filtering)
                    file_size = os.path.getsize(img_path) / (1024 * 1024)  # MB
                    if file_size > 5:  # Skip files larger than 5MB (reduced from 15MB)
                        continue
                    self.pairs.append((img_path, price))
                except:
                    continue
       
        print(f"Dataset created with {len(self.pairs)} image-price pairs")
       
        if len(self.pairs) == 0:
            raise ValueError("No valid image-price pairs found!")
   
    def __len__(self):
        return len(self.pairs)
   
    def __getitem__(self, idx):
        img_path, price = self.pairs[idx]
       
        try:
            with warnings.catch_warnings():
                warnings.simplefilter("ignore")
                image = Image.open(img_path).convert("RGB")
           
            if self.transform:
                image = self.transform(image)
           
            # Apply log transformation to price for stable training
            log_price = np.log(price)
           
            return image, torch.tensor([log_price], dtype=torch.float32)
       
        except Exception as e:
            print(f"Error loading image: {img_path}, error: {e}")
            # Return default values
            default_image = torch.zeros((3, 224, 224))
            default_price = torch.tensor([np.log(100.0)], dtype=torch.float32)
            return default_image, default_price

# ============================
# STEP 3: MODEL ARCHITECTURE
# ============================

class AirbnbPricePredictor(nn.Module):
    def __init__(self):
        super().__init__()
        # Use ResNet18 as backbone
        self.backbone = models.resnet18(weights="IMAGENET1K_V1")
       
        # Get the number of features from the backbone
        num_features = self.backbone.fc.in_features
       
        # Replace final layer with regression head
        self.backbone.fc = nn.Sequential(
            nn.Linear(num_features, 512),
            nn.ReLU(),
            nn.BatchNorm1d(512),
            nn.Dropout(0.3),  # Reduced dropout
            nn.Linear(512, 256),
            nn.ReLU(),
            nn.BatchNorm1d(256),
            nn.Dropout(0.2),  # Reduced dropout
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Dropout(0.1),  # Reduced dropout
            nn.Linear(128, 1)
        )
   
    def forward(self, x):
        return self.backbone(x)

# ============================
# STEP 4: TRAINING FUNCTIONS
# ============================

def evaluate_model(model, data_loader, device, criterion):
    """Evaluate model on validation/test set"""
    model.eval()
    total_loss = 0.0
    all_predictions = []
    all_actuals = []
   
    with torch.no_grad():
        for images, prices in tqdm(data_loader, desc="Evaluating"):
            images = images.to(device)
            prices = prices.to(device)
           
            outputs = model(images)
            loss = criterion(outputs.squeeze(), prices.squeeze())
            total_loss += loss.item()
           
            all_predictions.extend(outputs.cpu().numpy())
            all_actuals.extend(prices.cpu().numpy())
   
    avg_loss = total_loss / len(data_loader)
   
    # Convert back to actual prices for MAE calculation
    predictions = np.array(all_predictions).flatten()
    actuals = np.array(all_actuals).flatten()
   
    # Clip predictions to prevent exp() explosion
    predictions = np.clip(predictions, -2, 8)  # Roughly $7 to $3000 range
    actuals = np.clip(actuals, -2, 8)
   
    pred_prices = np.exp(predictions)
    actual_prices = np.exp(actuals)
   
    # Handle any remaining inf/nan values
    valid_mask = np.isfinite(pred_prices) & np.isfinite(actual_prices)
    if np.sum(valid_mask) > 0:
        mae = np.mean(np.abs(pred_prices[valid_mask] - actual_prices[valid_mask]))
    else:
        mae = float('inf')
   
    return avg_loss, mae

def train_model(model, train_loader, val_loader, device, epochs=30, lr=0.0005):
    """Train the model with early stopping and mixed precision"""
    os.makedirs("checkpoints", exist_ok=True)
   
    model = model.to(device)
    optimizer = optim.AdamW(model.parameters(), lr=lr, weight_decay=0.01)
    criterion = nn.MSELoss()
    scheduler = ReduceLROnPlateau(optimizer, mode="min", patience=3, factor=0.5)
   
    # Mixed precision training for speed
    scaler = torch.cuda.amp.GradScaler()
   
    best_val_loss = float("inf")
    train_losses, val_losses = [], []
    patience_counter = 0
    patience_limit = 7
   
    for epoch in range(epochs):
        # Training phase
        model.train()
        train_loss = 0.0
        train_batches = 0
       
        progress_bar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}")
       
        for images, prices in progress_bar:
            images = images.to(device, non_blocking=True)
            prices = prices.to(device, non_blocking=True)
           
            optimizer.zero_grad()
           
            # Mixed precision forward pass
            with torch.cuda.amp.autocast():
                outputs = model(images)
                loss = criterion(outputs.squeeze(), prices.squeeze())
           
            # Mixed precision backward pass
            scaler.scale(loss).backward()
            scaler.unscale_(optimizer)
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            scaler.step(optimizer)
            scaler.update()
           
            train_loss += loss.item()
            train_batches += 1
           
            progress_bar.set_postfix({"loss": f"{loss.item():.4f}"})
       
        avg_train_loss = train_loss / train_batches
       
        # Validation phase
        avg_val_loss, val_mae = evaluate_model(model, val_loader, device, criterion)
       
        train_losses.append(avg_train_loss)
        val_losses.append(avg_val_loss)
       
        print(f"\nEpoch {epoch+1}/{epochs}:")
        print(f"  Train Loss: {avg_train_loss:.4f}")
        print(f"  Val Loss: {avg_val_loss:.4f}")
        print(f"  Val MAE: ${val_mae:.2f}")
       
        # Save best model
        if avg_val_loss < best_val_loss:
            best_val_loss = avg_val_loss
            torch.save({
                "model_state_dict": model.state_dict(),
                "optimizer_state_dict": optimizer.state_dict(),
                "epoch": epoch,
                "best_val_loss": best_val_loss,
            }, "checkpoints/best_model.pth")
            print("   New best model saved!")
            patience_counter = 0
        else:
            patience_counter += 1
       
        # Early stopping
        if patience_counter >= patience_limit:
            print(f"\n Early stopping triggered at epoch {epoch+1}")
            break
       
        scheduler.step(avg_val_loss)
   
    # Plot training history
    plot_training_history(train_losses, val_losses)
    return model

def plot_training_history(train_losses, val_losses):
    """Plot training and validation loss"""
    epochs = len(train_losses)
    plt.figure(figsize=(10, 6))
    plt.plot(range(1, epochs + 1), train_losses, label="Train Loss", marker='o')
    plt.plot(range(1, epochs + 1), val_losses, label="Validation Loss", marker='s')
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.title("Training and Validation Loss")
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.savefig("training_history.png", dpi=300, bbox_inches='tight')
    plt.show()

# ============================
# STEP 5: INFERENCE FUNCTION
# ============================

def predict_listing_price(model, listing_id, image_root, device):
    """Predict price for a specific listing"""
    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])
   
    model.eval()
    predictions = []
   
    listing_dir = os.path.join(image_root, str(listing_id))
    if not os.path.exists(listing_dir):
        print(f"No images found for listing {listing_id}")
        return None
   
    # Get all images
    image_extensions = ["*.jpg", "*.jpeg", "*.png", "*.webp"]
    images = []
    for ext in image_extensions:
        images.extend(glob.glob(os.path.join(listing_dir, ext)))
   
    if not images:
        print(f"No valid images found for listing {listing_id}")
        return None
   
    with torch.no_grad():
        for img_path in images:
            try:
                image = Image.open(img_path).convert("RGB")
                image = transform(image).unsqueeze(0).to(device)
               
                output = model(image)
                # Convert log price back to actual price
                actual_price = np.exp(output.item())
                predictions.append(actual_price)
            except Exception as e:
                print(f"Error processing {img_path}: {e}")
                continue
   
    if predictions:
        avg_prediction = np.mean(predictions)
        print(f"Processed {len(predictions)} images for listing {listing_id}")
        return avg_prediction
    else:
        return None

# ============================
# STEP 6: MAIN TRAINING FUNCTION
# ============================

def main():
    print("🏠 AIRBNB PRICE PREDICTION CNN")
    print("=" * 50)
   
    # Configuration
    CSV_PATH = "/kaggle/input/listings-av/processed_listings.csv"  # Update this path
    IMAGE_ROOT = "/kaggle/input/airbnb-images-berlin"
    BATCH_SIZE = 64  # Increased from 32
    EPOCHS = 30
    LEARNING_RATE = 0.0001  # Reduced learning rate
   
    # Set device
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"Using device: {device}")
   
    # Step 1: Create matched dataset
    matched_df = create_matched_dataset(CSV_PATH, IMAGE_ROOT)
    if matched_df is None:
        print("Failed to create matched dataset!")
        return
   
    # Step 2: Define transforms
    train_transform = transforms.Compose([
        transforms.Resize((256, 256)),
        transforms.RandomCrop(224),
        transforms.RandomHorizontalFlip(),
        transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2),
        transforms.RandomRotation(10),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])
   
    val_transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])
   
    # Step 3: Create datasets
    print("\nCreating datasets...")
   
    # Split data into train/val/test (80/10/10)
    shuffled_df = matched_df.sample(frac=1, random_state=42).reset_index(drop=True)
    total_size = len(shuffled_df)
    train_size = int(0.8 * total_size)
    val_size = int(0.1 * total_size)
   
    train_df = shuffled_df[:train_size]
    val_df = shuffled_df[train_size:train_size + val_size]
    test_df = shuffled_df[train_size + val_size:]
   
    # Create datasets
    train_dataset = AirbnbDataset(train_df, IMAGE_ROOT, train_transform)
    val_dataset = AirbnbDataset(val_df, IMAGE_ROOT, val_transform)
    test_dataset = AirbnbDataset(test_df, IMAGE_ROOT, val_transform)
   
    # Create data loaders with more workers
    train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True,
                            num_workers=4, pin_memory=True, persistent_workers=True)
    val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False,
                          num_workers=4, pin_memory=True, persistent_workers=True)
    test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False,
                           num_workers=4, pin_memory=True, persistent_workers=True)
   
    print(f" Train samples: {len(train_dataset):,}")
    print(f" Validation samples: {len(val_dataset):,}")
    print(f" Test samples: {len(test_dataset):,}")
   
    # Step 4: Create and train model
    print("\nCreating model...")
    model = AirbnbPricePredictor()
   
    # Multi-GPU support
    if torch.cuda.device_count() > 1:
        print(f"Using {torch.cuda.device_count()} GPUs for parallel training!")
        model = nn.DataParallel(model)
   
    print("\n🚀 Starting training...")
    trained_model = train_model(model, train_loader, val_loader, device,
                              epochs=EPOCHS, lr=LEARNING_RATE)
   
    # Step 5: Final evaluation
    print("\n FINAL EVALUATION ON TEST SET")
    print("=" * 40)
   
    # Load best model
    checkpoint = torch.load("checkpoints/best_model.pth")
    model.load_state_dict(checkpoint["model_state_dict"])
    print(f" Best model loaded from epoch {checkpoint['epoch'] + 1}")
   
    # Test evaluation
    criterion = nn.MSELoss()
    test_loss, test_mae = evaluate_model(model, test_loader, device, criterion)
    print(f" Final Test Loss: {test_loss:.4f}")
    print(f" Final Test MAE: ${test_mae:.2f}")
   
    # Step 6: Sample predictions
    print("\n SAMPLE PREDICTIONS")
    print("=" * 30)
   
    sample_listings = test_df.head(10)
    successful_predictions = 0
    total_error = 0
   
    for _, row in sample_listings.iterrows():
        listing_id = row["listing_id"]
        actual_price = row["price"]
       
        predicted_price = predict_listing_price(model, listing_id, IMAGE_ROOT, device)
       
        if predicted_price:
            error = abs(predicted_price - actual_price)
            error_pct = (error / actual_price) * 100
           
            print(f"\nListing {listing_id}:")
            print(f"  Actual: ${actual_price:.2f}")
            print(f"  Predicted: ${predicted_price:.2f}")
            print(f"  Error: ${error:.2f} ({error_pct:.1f}%)")
           
            successful_predictions += 1
            total_error += error_pct
   
    if successful_predictions > 0:
        avg_error = total_error / successful_predictions
        print(f"\n Average prediction error: {avg_error:.1f}%")
   
    print(f"\n Training completed! Model saved at: checkpoints/best_model.pth")

if __name__ == "__main__":
    main()
```
